```python
import jax
import jax.numpy as jnp
import time
import os

# Verify device
print(jax.devices())

def benchmark(fn, *args, n_runs=10):
    # Warmup
    fn(*args).block_until_ready()
    times = []
    for _ in range(n_runs):
        t0 = time.time()
        fn(*args).block_until_ready()
        times.append(time.time() - t0)
    return min(times), sum(times)/len(times)

@jax.jit
def matmul(x, y):
    return jnp.dot(x, y)

sizes = [1024, 2048, 4096]
print(f"\n{'Size':<12} {'Min(s)':<12} {'Avg(s)':<12}")
print("-" * 36)
for s in sizes:
    x = jnp.ones((s, s))
    mn, avg = benchmark(matmul, x, x)
    print(f"{s}x{s:<8} {mn:<12.4f} {avg:<12.4f}")
```
